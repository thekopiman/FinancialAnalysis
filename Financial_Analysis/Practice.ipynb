```python
common_dates = {'2009-12-04', '2009-12-07', '2009-12-08', '2009-12-09', '2009-12-10', '2009-12-11', '2009-12-14', '2009-12-15'}

new_df = amazon_daily2[amazon_daily2.date.isin(common_dates)].filter(items=["date", "close"])
new_df
```

```python
returns = [0]
for i in range(len(new_df.close.values)):
    if i != 0:
        returns.append(new_df.close.values[i]/new_df.close.values[i-1] -1)

new_df['returns'] = returns
new_df
```
