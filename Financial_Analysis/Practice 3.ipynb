```python
dataframe1 = pd.read_excel('Everything.xlsx')
dataframe1
```

```python
np_array = dataframe1.to_numpy()
```

```python
for column in range(3):
    for row in range(1750):
        i = 1
        if np.isnan(np_array[row, column]):
            while True:
                try:
                    if np.isnan(np_array[row+i, column]):
                        i+=1
                    else:
                        np_array[row, column] = np_array[row+i, column]
                        np_array[row+i, column] = np.nan
                        break
                except IndexError:
                    break

```
